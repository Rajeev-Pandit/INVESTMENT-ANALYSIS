Checkpoint 2: Funding Type Analysis

-- Start of coding steps to identify most suitable fund for the company Spark Funds

```python
master_frame['raised_amount_usd'].describe()
```

Here there is a difference between mean value (approx. 10.43M) and median value (approx. 1.68M). There's a significant variation in the 25th and the 75th percentile as well (0.32M to 7M). Thus, imputing NaNs with mean/median seems quite biased, and so I would remove the NaNs.  

```python
# removing NaNs in column raised_amount_used
master_frame = master_frame[~np.isnan(master_frame['raised_amount_usd'])]

round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

```python
#All Nans removed from column raised_amount_usd
master_frame.head(5)
```

Next grouping of the coulmn funding_round_type will be carried out to group same types of funds together 

```python
# I will group the master dataframe column funding_round_type by fund types
fund_by_type = master_frame.groupby('funding_round_type')
fund_by_type
```

```python
# sorting of the average (using mean) fund values, million USD, in descending order:
fund_by_type.raised_amount_usd.mean().div(1000000).apply(lambda x: round(x,2)).sort_values(ascending = False)
```

```python
# sorting of the average (using median) fund values, million USD, in descending order:
fund_by_type.raised_amount_usd.median().div(1000000).apply(lambda x: round(x,3)).sort_values(ascending = False)
```

From the above average fund values list (based on mean and median), it can be said that venture fund (among the venture, angel, seed and private_equity funds) average value  is the one which lies between 5 million to 15 million USD and hence venture fund is the most suitable fund type for the company Spark Funds.  

-- End of the coding steps for identifying most suitable fund
